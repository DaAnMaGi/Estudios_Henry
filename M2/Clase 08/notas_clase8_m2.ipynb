```python
import pymysql # Importar módulo
```

```python
# Conectar python con el servidorde workbench
conexion = pymysql.connect(
    host = "localhost",
    user = "root",
    passwd= "Alaver_456",
    port=3306,
    db = "clase_5"
)

cursor = conexion.cursor()
```

Traer información

```python
cursor.execute("select * from alumnos")
alumnos = cursor.fetchall()

for alumno in alumnos:
    print(alumno[0:4])
```

```python
cursor.execute("select * from alumnos")
alumnos = cursor.fetchone() #Trae el primero

for alumno in alumnos:
    print(alumno)
```

```python
cursor.execute("select * from alumnos")
alumnos = cursor.fetchmany(5)

for alumno in alumnos:
    print(alumno)
```

```python
cursor.execute("select * from alumnos")
for alumno in cursor:
    print(alumno)
```

```python
cursor.execute('''select idAlumno,nombre,apellido 
               from alumnos
               limit 5''')
for alumno in cursor:
    print(alumno)
```

```python
cursor.execute('''select idAlumno,concat(nombre," ",apellido) 
               from alumnos
               limit 5''')
for alumno in cursor:
    print(alumno)
```

Insertar datos

```python
cursor.execute('''insert into alumnos
               values (181,"32547841","Pablo","Gonzalez","1998-06-06","2022-05-06",1243)
               ''' )
```

```python
# La inserción requiere un commit. Los delete y los update sí los hace de manera inmediata. 

conexion.commit()
```

```python
# Una vez se haya terminado, se recomienda cerrar la colección.

conexion.close()
```
