```python
def NumeroBinario(numero):
    '''
    Esta función recibe como parámetro un número entero mayor ó igual a cero y lo devuelve en su 
    representación binaria. Debe recibir y devolver un valor de tipo entero.
    En caso de que el parámetro no sea de tipo entero y mayor a -1 retorna nulo.
    '''
    if type(numero) != int:
        return None
    if numero < 0:
        return None
    if numero == 0:
        return 0
    
    x = numero
    binario = ""
    
    while x > 0:
        binario = str(x%2)+binario
        x = x//2
            
    return int(binario)
    
```

2) Convertir de decimal a binario las fracciones 1/2, 1/3, 1/4, 1/5, 1/6, 1/7, 1/8, 1/9. Luego analizar los resultados y observar qué particularidad se encuentra en los mismos. Se puede usar Python o una calculadora, lo importante es ver si hay algo que podemos notar...

Salida esperada:
* Fraccion 1 / 2 :  0.5  y En binario:  0.1
* Fraccion 1 / 3 :  0.3333333333333333  y En binario:  0.010101010101010101010101
* Fraccion 1 / 4 :  0.25  y En binario:  0.01
* Fraccion 1 / 5 :  0.2  y En binario:  0.001100110011001100110011
* Fraccion 1 / 6 :  0.16666666666666666  y En binario:  0.001010101010101010101010
* Fraccion 1 / 7 :  0.14285714285714285  y En binario:  0.001001001001001001001001
* Fraccion 1 / 8 :  0.125  y En binario:  0.001
* Fraccion 1 / 9 :  0.1111111111111111  y En binario:  0.000111000111000111000111

```python
def Fraccion_Binario (fraccion,k=18):
    '''Toma una fracción o un decimal y la convierte en un número binario, la fórmula toma dos datos: la fracción, y el número de cifras que se tomará del decimal. Si no se le da un número, tomará las primeras 18 cifras del decimal. 
    '''
    x = fraccion
    entero = int(x)
    decimal = x - entero
    binario = ""

    #Se transforma la parte entera en binario
    binario += str(NumeroBinario(entero)) + "."

    #Se transforma la parte decimal en binario
    while (decimal > 0) and (k):
        decimal *= 2
        bit = int(decimal)

        if bit ==1:
            decimal -= 1
            binario += "1"
        else: binario += "0"
        k -= 1

    return binario
    

```

```python
print(Fraccion_Binario(1/2))
print(Fraccion_Binario(1/3))
print(Fraccion_Binario(1/4))
print(Fraccion_Binario(1/5))
print(Fraccion_Binario(1/6))
print(Fraccion_Binario(1/7))
print(Fraccion_Binario(1/8))
print(Fraccion_Binario(1/9))
```

```python
for i in range(2,10):
    print(f"Fracción 1/{i}: {str(1/i)} y en binario: {Fraccion_Binario(1/i)}")
```
