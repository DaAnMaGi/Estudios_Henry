## Indicadores desarrollo humano

```python
filas = ["PIB (US$ a precios actuales)","Crecimiento del PIB (% anual)","Población entre 0 y 14 años de edad, total","Población de 65 años de edad y más, total","Población entre 15 y 64 años de edad, total"] # Se seleccionan las filas de datos que se consideran relevantes
dsh = desarrolloHumano[desarrolloHumano["Series Name"].isin(filas)]
dsh.drop(["Series Code"],axis=1,inplace=True)
dsh.head()
```

```python
nombre = dsh["Country Name"].unique()
codigo = dsh["Country Code"].unique()
zonas = pd.DataFrame([nombre,codigo]).transpose()
zonas.rename(columns={0:"Nombre Pais",1:"CodigoPais"},inplace=True)
```

```python
zonas
```

```python
zonas.to_csv("./Bases_limpias/zonas.csv")
```

```python
dsh.drop(["Country Name"],axis=1,inplace=True)
dsh.head()
```

```python
años = [2000]
i = 0
while len(años) < 20:
    i += 1
    años.append(2000+i)
```

```python
USA = dsh[dsh["Country Code"]=="USA"]
USA_T = USA.transpose()
USA_T = USA_T.set_axis(["PIB","Crecimiento PIB","Población mayor 65 años","Población entre 0 y 14 años","Población entre 15-64 años"],axis=1)
USA_T.insert(0,"CodigoPais","USA",allow_duplicates=True)
USA_T = USA_T.drop("Series Name")
USA_T = USA_T.drop("Country Code")
USA_T.insert(1,"Año",años,allow_duplicates=False)

USA_T.head()
```

```python
# EUU JPN WLD
EUU = dsh[dsh["Country Code"]=="EUU"]
EUU_T = EUU.transpose()
EUU_T = EUU_T.set_axis(["PIB","Crecimiento PIB","Población mayor 65 años","Población entre 0 y 14 años","Población entre 15-64 años"],axis=1)
EUU_T.insert(0,"CodigoPais","EUU",allow_duplicates=True)
EUU_T = EUU_T.drop("Series Name")
EUU_T = EUU_T.drop("Country Code")
EUU_T.insert(1,"Año",años,allow_duplicates=False)

EUU_T.head()
```

```python
JPN = dsh[dsh["Country Code"]=="JPN"]
JPN_T = JPN.transpose()
JPN_T = JPN_T.set_axis(["PIB","Crecimiento PIB","Población mayor 65 años","Población entre 0 y 14 años","Población entre 15-64 años"],axis=1)
JPN_T.insert(0,"CodigoPais","JPN",allow_duplicates=True)
JPN_T = JPN_T.drop("Series Name")
JPN_T = JPN_T.drop("Country Code")
JPN_T.insert(1,"Año",años,allow_duplicates=False)

JPN_T.head()
```

```python
WLD = dsh[dsh["Country Code"]=="WLD"]
WLD_T = WLD.transpose()
WLD_T = WLD_T.set_axis(["PIB","Crecimiento PIB","Población mayor 65 años","Población entre 0 y 14 años","Población entre 15-64 años"],axis=1)
WLD_T.insert(0,"CodigoPais","WLD",allow_duplicates=True)
WLD_T = WLD_T.drop("Series Name")
WLD_T = WLD_T.drop("Country Code")
WLD_T.insert(1,"Año",años,allow_duplicates=False)

WLD_T.head()
```

```python
dsh_limpio = pd.concat([USA_T,EUU_T,JPN_T,WLD_T],ignore_index=True)
dsh_limpio = dsh_limpio.replace("..",None)
dsh_limpio
```

```python
dsh_limpio.to_csv("./Bases_limpias/indicadores_dsh.csv")
```

## Ventas juegos

```python
ventas
```

```python
vs = ventas
```

```python
i = 0
indices = []
while len(indices) < len(vs["Name"]):
    i += 1
    indices.append(i)
```

```python
vs.insert(0,"Id",indices,allow_duplicates=False)
vs
```

```python
vs[vs["Global_Sales"].isna()]
```

```python
juegos_ventas = vs.loc[:,["Id","NA_Sales","EU_Sales","JP_Sales","Other_Sales","Global_Sales"]]

# Obtención de valores GLobal_Sales que son nulos
juegos_ventas["Global_Sales"].fillna(sum([juegos_ventas["NA_Sales"],juegos_ventas["EU_Sales"],juegos_ventas["JP_Sales"],juegos_ventas["Other_Sales"]]),inplace=True)
juegos_ventas.head()
```

```python
# Obtención de valores GLobal_Sales que siguen siendo nulos
juegos_ventas[juegos_ventas["Global_Sales"].isna()].fillna(0,inplace=True)
juegos_ventas[juegos_ventas["Global_Sales"] == 0].replace(0,sum([juegos_ventas["NA_Sales"],juegos_ventas["EU_Sales"],juegos_ventas["JP_Sales"],juegos_ventas["Other_Sales"]]),inplace=True)
```

```python
juegos_ventas[juegos_ventas["Global_Sales"].isna()]
```

```python
juegos_ventas[(juegos_ventas["NA_Sales"].isnull()) & (juegos_ventas["Global_Sales"] > sum([juegos_ventas["EU_Sales"],juegos_ventas["JP_Sales"],juegos_ventas["Other_Sales"]]))].fillna((juegos_ventas["Global_Sales"] - sum([juegos_ventas["EU_Sales"],juegos_ventas["JP_Sales"],juegos_ventas["Other_Sales"]])),inplace=True)
```

```python
juegos_ventas[(juegos_ventas["NA_Sales"].isnull()) & (juegos_ventas["Global_Sales"] > sum([juegos_ventas["EU_Sales"],juegos_ventas["JP_Sales"],juegos_ventas["Other_Sales"]]))]
```
